```python
td.drop(['Sex','Pclass','Name'], axis=1, inplace=True)
```

```python
td.head(20)
```

### Train Data

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report as creport
from sklearn.metrics import confusion_matrix as conmatrix
from sklearn.metrics import accuracy_score as accscore
```

```python
X = td.drop("Survived", axis=1)
```

```python
y = td["Survived"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
creport(y_test, predictions)
```

```python
conmatrix(y_test, predictions)
```

```python
accscore(y_test, predictions)
```
